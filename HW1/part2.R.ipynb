```r
bounding <- function(dataIN){
    cubic <- matrix(,nrow(dataIN),400) 
    for (i in 1:nrow(dataIN)){
        cubic28 <- matrix(dataIN[i, ],28,28)
        for (j in 1:28){
            if (Reduce("+",cubic28[j,])>0){
                lowerbound <- j
                break
            }
        }
        for (j in 28:1){
            if (Reduce("+",cubic28[j,])>0){
                upperbound <- j
                break
            }
        }
        for (j in 1:28){
            if (Reduce("+",cubic28[,j])>0){
                leftbound <- j
                break
            }
        }
        for (j in 28:1){
            if (Reduce("+",cubic28[,j])>0){
                rightbound <- j
                break
            }
        }
        cubic20 <- cubic28[lowerbound:upperbound, leftbound:rightbound]
        cubic20 <- apply(resizeImage(as.matrix.data.frame(cubic20), 20, 20, method = "bilinear"), 1, function(x)x*255)
        cubic[i, ] <- cubic20
    }
    return (cubic)
}
```

```r
train.cubic = bounding(train.feature)
val.cubic = bounding(val.feature)
```

```r
colnames(train.cubic) <- colnames(train.feature)[1:400]
colnames(val.cubic) <- colnames(train.feature)[1:400]
```

```r
model3 <- train(x = train.cubic, y = train.label, method = 'naive_bayes', trControl=trainControl(method='cv', number=10))
```

```r
val.pred3 <- predict(model3, newdata = val.cubic)
cmp = confusionMatrix(data = val.pred3, val.label)
cmp
```

```r
model4 <- textmodel_nb(as.dfm(train.cubic), as.factor(train.label), distribution = 'Bernoulli')
```

```r
val.pred4 <- predict(model4, newdata = as.dfm(val.cubic))
cmp = confusionMatrix(data = val.pred4, val.label)
cmp
```

```r
randomForestFunc = function(X, Y, trees, depth){
    forestModel = h2o.randomForest(y = "label", training_frame = X, validation_frame = Y, ignore_const_cols = F, ntrees = trees, max_depth = depth, score_each_iteration = T) 
    print(h2o.hit_ratio_table(forestModel, valid = T)[1,2])
}
```

```r
h2o.init()
```

```r
frame1 = dataTrain[,2:786]
frame1$label <- as.factor(frame1$label)
frame2 = dataVal
frame2$label <- as.factor(frame2$label)
```

```r
randomForestFunc(as.h2o(frame1), as.h2o(frame2), 10, 4);
randomForestFunc(as.h2o(frame1), as.h2o(frame2), 30, 4);
randomForestFunc(as.h2o(frame1), as.h2o(frame2), 10, 16);
randomForestFunc(as.h2o(frame1), as.h2o(frame2), 30, 16);
```

```r
frame3 <- as.h2o(list(frame1[,1], bounding(dataTrain[ ,3:786])))
frame4 <- as.h2o(list(frame2[,1], bounding(dataVal[ ,2:785])))
```

```r
names(frame3)[1] = c("label")
names(frame4)[1] = c("label")
```

```r
randomForestFunc(as.h2o(frame3), as.h2o(frame4), 10, 4);
randomForestFunc(as.h2o(frame3), as.h2o(frame4), 30, 4);
randomForestFunc(as.h2o(frame3), as.h2o(frame4), 10, 16);
randomForestFunc(as.h2o(frame3), as.h2o(frame4), 30, 16);
```

```r
bestForestModel = h2o.randomForest(y = "label", training_frame = as.h2o(frame3), validation_frame = as.h2o(frame4), ignore_const_cols = F, ntrees = 30, max_depth = 16, score_each_iteration = T)
```
